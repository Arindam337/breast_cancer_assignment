```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
```

```python
df=df.drop(['id'],axis=1)
```

```python
df
```

```python
df['diagnosis'] = df['diagnosis'].replace("M", 1) 
df['diagnosis'] = df['diagnosis'].replace("B", 0)
```

```python
df
```

```python
df['radius_mean']=df['radius_mean'].fillna(df['radius_mean'].median())
df['area_mean']=df['area_mean'].fillna(df['area_mean'].median())
```

```python
df.isnull().sum()
```

```python
X=df.drop("diagnosis",axis=1)
Y=df[["diagnosis"]]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.30,random_state=1)
```

```python
reg_model=LogisticRegression()
reg_model.fit(X_train,Y_train)
```

```python
reg_model.score(X_train,Y_train)
```

```python
reg_model.fit(X_test,Y_test)
reg_model.score(X_test,Y_test)
```

```python
predictions=reg_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,predictions)
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(Y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test,predictions)
```

```python
cm=metrics.confusion_matrix(Y_test,predictions,labels=[1,0])
```

```python
df_cm=pd.DataFrame(cm,index=[i for i in ["1","0"]],
                  columns=[i for i in ["Predict 1","Predict 0"]])
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(df_cm,annot=True,fmt='g')
```

#### The heat map shows that the model can predict 104 true benign and 58 malignant tumor correctly out of a total sample of (58+5+4+104)=171 samples.Hence the efficiency of the model is (162/171)*100 = 94.73%. However, it predicts 4 samples as malignant which are benign actually and 5 samples benign which are malignant in reality.
